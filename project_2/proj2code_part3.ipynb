```python
import pandas as pd
```

```python
# Revisit the features picked out from Lasso.
lassofeatures = pd.read_csv('output/lasso_coefs.csv')
lassofeatures.drop(columns='Unnamed: 0', inplace=True) # Drop the unnamed column.
print(lassofeatures.shape)
lassofeatures[:15]
```

```python
# Rank lasso features and rename to prepare for merge.
lassofeatures['lassorank'] = lassofeatures.index + 1
lassofeatures.rename(columns={'variable': 'features','coef': 'lasso_coef', 'abs_coef': 'lasso_abs_coef'}, inplace=True)
lassofeatures.head()
```

```python
# Revisit the features picked out from ridge.
ridgefeatures = pd.read_csv('output/ridge_coefs.csv')
ridgefeatures.drop(columns='Unnamed: 0', inplace=True) # Drop the unnamed column.
print(ridgefeatures.shape)
ridgefeatures[:15]
```

```python
# Rank ridge features and rename to prepare for merge.
ridgefeatures['ridgerank'] = ridgefeatures.index + 1
ridgefeatures.rename(columns={'variable': 'features','coef': 'ridge_coef', 'abs_coef': 'ridge_abs_coef'}, inplace=True)
ridgefeatures.head()
```

```python
# Combine lasso and ridge features.
comb_rank = lassofeatures.merge(ridgefeatures, how = 'left', on = 'features')
comb_rank['combrank'] = comb_rank['lassorank'] + comb_rank['ridgerank']
# Sort by combined rank to find top features.
comb_rank = comb_rank.sort_values(by = 'combrank')
comb_rank[:30]
```

I revisited both the lists and observed that they identified somewhat different features as top influences of saleprice. As such I combined them to get the best of both worlds.
