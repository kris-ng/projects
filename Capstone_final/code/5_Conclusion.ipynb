## Future Works

A few aspects can be refined for better model prediction: 

- neutral' dominant emotion class  

I'm not too happy that 'neutral' seems to have dominated the data despite doing an undersampling in pycaret. So on hindsight, I could have considered dropping this emotion class altogether so that my model has to force-classify into the emotions that range from negative to positive. Many a times when we fill in questionnaires, we tend to choose the 'neutral' answer too so this may be a similar trend - to assign to it when uncertain.

- word length  

I may even consider selecting only comments of a specified word length range so that I keep it uniform across the emotion classes. My opinion is that those on the extreme ends, whether very short or very long texts may end up having its classification distorted. Like perhaps a comment could have been a mix of emotions but ended up being classified based on more words displaying a particular emotion. 

- group comments by time zone  
If I have the means to, I would like to group the comments based on their time zone so that the analysis of whether positive/ negative comments are more common at different junctures of the day is more accurate an observation. As it is currently, all the comments are lumped together so regardless of which time zone they belong to, the observation may not be as reliable as I would have liked it to be. 

The model can be continually refined by feeding it with new comments that keep up with the current topic of interests and economical situations so that it better interprets the sentiments. Where resources allow, the model can predict a wider range of emotions to better cater to users' needs. 

If this model takes off, the project can be taken further by availing a user-initiated chat where users can choose to access at various points in time to check on themselves in order to take the appropriate actions to enhance their state of mental health.
